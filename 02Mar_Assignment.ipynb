# Answer2:
A scatter plot is a type of data visualization that shows the relationship between different variables. This data is shown by placing various data points between an x- and y-axis. Essentially, each of these data points looks “scattered” around the graph, giving this type of data visualization its name.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Plot of given data")
```

# Answser3:
Subplot divides the current figure into rectangular panes that are numbered row-wise. Each pane contains an axes. Subsequent plots are output to the current pane. subplot(m,n,p) creates an axes in the p -th pane of a figure divided into an m -by- n matrix of rectangular panes.

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x1,y1)
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x2,y2)
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x3,y3)
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x4,y4)
```

# Answer4:
A bar chart is used when you want to show a distribution of data points or perform a comparison of metric values across different subgroups of your data. From a bar chart, we can see which groups are highest or most common, and how other groups compare against the others. Bar graphs are used to compare things between different groups or to track changes over time.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(5,3))
plt.barh(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
```

# Answer5:
Box Plot: It is a type of chart that depicts a group of numerical data through their quartiles. It is a simple way to visualize the shape of our data. It makes comparing characteristics of data between categories very easy. Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
```

```python
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
```
